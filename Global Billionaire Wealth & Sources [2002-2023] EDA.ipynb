# Visualisations and Metrics

**1.Women and Men ratio on the dataset**

```python
gender_count = df.gender.value_counts().drop("M;F")
plt.figure(figsize=(10,5))
plt.title("Women and Men Ratio on the Dataset")
gender_count.plot(kind='bar')
```

```python
by_gender = df[(df['gender']!='Not known')]
by_gender = by_gender[(by_gender['gender']!='M;F')]
gender_rich = by_gender.groupby(['gender', 'time'])['annual_income'].agg({'count', 'sum'}).reset_index()
gender_rich.columns = ['Gender', 'Year', 'Count', 'Total_Wealth']
gender_rich.head(4)
```

```python
gender = pd.pivot_table(by_gender, index = 'time', columns = 'gender', 
                        values = 'annual_income', aggfunc = {'count', 'sum'})
gender.columns = ['Female_Count', 'Male_Count', 'Female_Wealth', 'Male_Wealth']
```

```python
gender['Female_n_perc'] = gender['Female_Count']/gender.loc[2002, 'Female_Count']*100
gender['Male_n_perc'] = gender['Male_Count']/gender.loc[2002, 'Male_Count']*100
gender['Female_W_perc'] = gender['Female_Wealth']/gender.loc[2002, 'Female_Wealth']*100
gender['Male_W_perc'] = gender['Male_Wealth']/gender.loc[2002, 'Male_Wealth']*100
gender = gender.reset_index()
gender.head(3)
```

```python
y = ['Female_n_perc', 'Male_n_perc','Female_W_perc', 'Male_W_perc']
# for legends
labels = ['Female Numbers Change', 'Male Numbers Change', 'Females Total Wealth Change', 'Males Total Wealth Change']
markers = ['o','v','*','.']
linestyles = ['-','--','-.',':']
colors = ["#b30000", "#7c1158", "#4421af", "#edbf33"]
plt.figure(figsize = (15,9))
for col, marker, linestyle, color, label in zip(y, markers,linestyles,colors, labels):
    plt.plot(gender['time'], gender[col], marker = marker, linestyle = linestyle, color = color, label = label)
plt.legend()
plt.xticks(gender['time'], rotation = 45);
plt.grid(axis = 'y')
plt.xlabel('Years', labelpad = 8,fontsize = 16)
plt.ylabel('Percentage', fontsize = 16)
plt.title('Males and Females Percentage Change in Count and Income', pad = 12, fontsize = 18);
plt.show()
```

Findings:

    1.The graph depicts that number of female rich people increased the most from 2002 to 2022 compare to number of male rich people .eventhough the number of female rich people increases but their total wealth decreases over the time period
    2. Another important fact that in 2009 , 2006 and 2020 , all showed downslide because of the financial crisis and corona virus.
    3.Also noticeable is the remarkable increase in all terms of over 600%.

**2.Total Income of Wealthies by Country**

```python
df['country'] = df['permanent_country'].str.split('-', expand=True)[0]
by_country = df[df['country']!='Not known']
top_7 = list(by_country.groupby('country')['annual_income'].sum().sort_values(ascending = False).index[:7])
top_countries = by_country[by_country['country'].isin(top_7)]
top_countries.head(3)
```

```python
fig = px.bar(top_countries, x="country", y="annual_income", color="country",
  animation_frame="time",log_y = True, range_y = [500000000, 120000000000])
fig.update_layout(title_text = 'Total Income of Wealthies by Country', title_x = 0.5)
fig.show()
```

Findings:
    
    1. USA is by far the most wealthiest country with total Amount Incomes of Top Rich People.
    2. China has shown tremendous growth from 2002 to 2022 compare to other other countries.
    3. It is noticable that By 2022 the total income of Russia is decreased(may be due to war).

**3.Periodical Rich People**

```python
every_year_rich = df['name'].value_counts().value_counts()[:-2]
every_year_rich = every_year_rich.sort_index().reset_index()
every_year_rich.columns = ['Yearly Period', 'Richs']
every_year_rich['Wealthy for'] = [a +' Year' if a == '1' else a+' Years' for a in every_year_rich['Yearly Period'].astype(str)]
every_year_rich.head(3)
```

```python
fig = px.pie(data_frame = every_year_rich, names = 'Wealthy for', values = 'Richs')
fig.update_traces(textposition='inside', textinfo='percent+label',)
fig.update_traces(hole=.4, hoverinfo="label+percent+name")
fig.update_layout(
    title_text="Periodical Rich People",title_x = 0.47,
    # Add annotations in the center of the donut pies.
    annotations=[dict(text='Duration', x=0.5, y=0.5, font_size=20, showarrow=False)])
fig.show()
```

Findings:
    
    1.Interestingly, 178 people (3.5%) were on the Top Rich List every single year from 2002 and 2022.

**4.Changes in Wealthy's Main Industries over 20 years**

```python
main_industry = df.main_industry.value_counts().head(10)
plt.figure(figsize=(10,5))
main_industry.plot(kind='bar')
```

```python
top_industries = list(df['main_industry'].value_counts().index[:10])
by_industry = df[df['main_industry'].isin(top_industries)]
grouped_indust = pd.pivot_table(by_industry, index = 'time', columns = 'main_industry', values = 'annual_income', aggfunc = 'sum')
grouped_indust.head(21)
```

```python
fig = px.line(grouped_indust.reset_index(), x = 'time', y = grouped_indust.columns[1:])
fig.update_layout(title_text = "Changes in Wealthy's Main Industries over 20 years", title_x = 0.43)
fig.show()
```

Findings:

    1.Technology, Fashion, Finance Industries made up the most Wealthy People's Income.
